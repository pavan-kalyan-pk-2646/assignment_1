## ASSIGNMENT 1

```python
import pandas as pd
# Load Uber dataset
uber_data = pd.read_csv("C:/Users/namei/Downloads/datasets-main/datasets-main/Uber_Drives_2016.csv")
uber_data.head()  # Display first 5 rows
```

##      ASSIGNMENT 1 Data Exploration

```python
# Shape of the dataset
print("Rows and columns:", uber_data.shape)

# Info about dataset
print(uber_data.info())

# Column names
print("Columns:", uber_data.columns)

# Unique categories in 'CATEGORY'
print("Unique categories:", uber_data['CATEGORY*'].unique())

# Check null values in 'PURPOSE'
print("Null values in PURPOSE*:", uber_data['PURPOSE*'].isnull().sum())

# Display rows with any missing values
null_rows = uber_data[uber_data.isnull().any(axis=1)]
print("Rows with missing values:\n", null_rows)

# Rename all columns in uppercase
uber_data.columns = [col.upper() for col in uber_data.columns]
print("Columns after renaming:", uber_data.columns)
```

## ASSIGNMENT 2 Data Filtering & Transformation

```python


# --- 1. Filter rides where CATEGORY == 'Business' ---
business_rides = uber_data[uber_data['CATEGORY*'] == 'Business'].copy()
print("Business rides:\n", business_rides.head())

# --- 2. Top 5 rides with longest distance ---
top5_distance = uber_data.nlargest(5, 'MILES*').copy()
print("Top 5 rides by distance:\n", top5_distance)

# --- 3. Replace missing PURPOSE values with "Not Specified" ---
uber_data['PURPOSE'] = uber_data['PURPOSE*'].fillna("Not Specified")
print("Updated PURPOSE column:\n", uber_data['PURPOSE*'].head(10))

# --- 4. Convert START_DATE and END_DATE to datetime safely ---
uber_data['START_DATE'] = pd.to_datetime(uber_data['START_DATE*'], errors='coerce')
uber_data['END_DATE'] = pd.to_datetime(uber_data['END_DATE*'], errors='coerce')

# Drop rows where conversion failed and create a copy
uber_data = uber_data.dropna(subset=['START_DATE*', 'END_DATE*']).copy()

# --- 5. Create TRIP_DURATION column ---
uber_data['TRIP_DURATION'] = uber_data['END_DATE'] - uber_data['START_DATE']
print("Trip duration sample:\n", uber_data[['START_DATE*','END_DATE*','TRIP_DURATION']].head())

# --- 6. Sort trips by distance in descending order ---
uber_sorted = uber_data.sort_values(by='MILES*', ascending=False).copy()
print("Trips sorted by distance:\n", uber_sorted.head())
```
